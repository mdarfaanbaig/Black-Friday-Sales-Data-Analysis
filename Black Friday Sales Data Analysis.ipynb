# Analyzing Gender

```python
data = pd.DataFrame({'Ratio': [len(df[df['Gender'] == 'M']),len(df[df['Gender'] == 'F'])]}, index = ['Male', 'Female'])
```

```python
data
```

```python
data.plot.pie(y = 'Ratio' , figsize = (5,5) , autopct = "%.1f")
# autopct is used to label the wedges with their numerical value in pie chart
```

```python
# more efficient way to code
df.groupby('Gender').size().plot(kind = 'pie',autopct = "%.1f",title = 'Gender Ratio', figsize = (5,5))
```

```python
df.groupby('Gender').size().plot(kind = 'bar', figsize=(5,5))
```

# Combining Gender And Marital Status

```python
l = []
for i in range(len(df)):
    l.append(df['Gender'][i] +"_"+ str(df['Marital_Status'][i]))

df['MaritalGender'] = l
```

```python
df.head()
```

```python
sns.countplot(x=df['MaritalGender'])
```

```python
sns.countplot(x=df['Age'] , hue = df['MaritalGender'])
```

```python
sns.countplot(x=df['Product_Category_1'] , hue = df['MaritalGender'])
```

```python
sns.countplot(x=df['Stay_In_Current_City_Years'] , hue = df['MaritalGender'])
```

```python
sns.countplot(x=df['City_Category'] , hue = df['MaritalGender'] )
```

```python
# Unmaried Male are spending the most on Black Friday Sales
# Then comes Married men and Unmarried Female and then Married female
# Priority Order : Unmaried Male > Married men > Unmarried Female > Married female
```
